```python
# Get all stats of a YT Channel

api_service_name = "youtube"
api_version = "v3"

youtube = build(
    api_service_name, api_version, developerKey=api_key)

request = youtube.channels().list(
    part="snippet,contentDetails,statistics",
    id="UC_x5XG1OV2P6uZZ5FSM9Ttw"
)

response = request.execute()


print(json.dumps(response, indent=4))
```

```python
# Define get_channel_stats function
def get_channel_stats(channel_id):
    request = youtube.channels().list(
        part="snippet,contentDetails,statistics",
        id=channel_id
    )
    response = request.execute()
    items = response["items"][0]
    channel_data = {
        "channel_name": items['snippet']['title'],
        "subscribers": items['statistics']['subscriberCount'],
        "views": items["statistics"]['viewCount'],
        "total_videos": items["statistics"]['videoCount'],
        "playlist_Id": items["contentDetails"]["relatedPlaylists"]["uploads"]
    }
    return channel_data
    # return pd.DataFrame([channel_data])
    
    
```

```python
# Get Info of a channel based on its id
channel_stats = get_channel_stats("UC_x5XG1OV2P6uZZ5FSM9Ttw")
channel_stats
```

```python
def get_channel_id_by_name(channel_name):
    request = youtube.search().list(
        part="snippet",
        q=channel_name,  # Search query for channel name
        type="channel",  # Only search for channels
        maxResults=3  #
    )
    response = request.execute()

    # Debug: print the response to inspect the structure
    if 'items' in response:
        channels = response['items']
        # Get channel ids by going through the given channel items
        channels_ids = [channel['snippet']['channelId'] for channel in channels]
        # return channels_ids # uncomment this if you want just ids 
        # get channel info by passing it to my get_channel_stats function
        channels_info = [get_channel_stats(id) for id in channels_ids]
        return channels_info
    else:
        return []

# Example usage
channel_name = "DanTDM"
responses = get_channel_id_by_name(channel_name)
print(responses)
```

```python
def get_channel_uploads(channel_id):
    request = youtube.playlistItems().list(
        part="snippet,contentDetails",
        playlistId=channel_id
    )
    response = request.execute()
    return json.dumps(response, indent=4)
```
